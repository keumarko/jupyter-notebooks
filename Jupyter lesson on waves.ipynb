# The mathematical representation of a wave

This lesson is the final part of your classes that deal with the concept of waves. 
A wave is a periodic (something that repeats itself regularly) phenomenon that propagates through space. 
As such, it can be represented by a trigonometric function such as sin(x) or cos(x). 
These functions look like this:

```python
import numpy
from matplotlib import pyplot
x = numpy.linspace(0,10*numpy.pi, 10**3)
y = numpy.cos(x)
z = numpy.sin(x)
```

```python
pyplot.plot(x,y)
pyplot.plot(x,z)
pyplot.title('The cosine and sine functions')
```

We could use either of these functions to represent our wave as the only difference between the two is that they are "out of phase" - meaning that they are at different points of their cycle when we look at them. The phase "difference" between the two functions is pi/2 (which is why sin(x+ pi/2) = cos(x)). We shall arbitrarily be looking at the sine function in this lesson.

## Amplitude of a wave

The amplitude of a sine wave varies between -1 and 1. That means that we can get any wave amplitude we wish by multiplying the function sin(x) by this amplitude. For example:

```python
amplitude = 6
y = amplitude * numpy.cos(x)
pyplot.plot(x,y)
pyplot.title('The amplitude of the sine wave has now been changed')
pyplot.xlabel('Time')
pyplot.ylabel('Amplitude')
```

You can try changing the amplitude of the sine wave yourself in the cell above (the one that states amplitude = 4) and run the code to see how the scale changes to match your input.

## Frequency of a wave

Now, remember that a prime characteristic of a wave is its frequency, i.e the number of waves that pass a fixed point every second. This can also be modelled mathematically by using something called its angular frequency omega, where omega = 2*pi*frequency

```python
frequency = 2
omega = 2*numpy.pi*frequency
t = numpy.linspace(0,2,10**3)
y = amplitude*numpy.cos(omega*t)
pyplot.plot(t,y)
pyplot.title("In this graph we can change the frequency of the wave")
pyplot.xlabel("Time")
pyplot.ylabel("Amplitude")
```

By changing the frequency in the box above you will observe that the number of waves per second also changes

## Phase of a wave

The phase of a wave was briefly discussed earlier and tells us how far in its cycle it is. The letter phi is usually used for the phase. One complete period of a wave is mathematically similar to a full rotation around a circle, so one cycle is like a rotation around an angle of 2 pi (as counted in radians). This means that if we change the phase of a wave by 2 pi, it will in fact look exactly the same as previously, since we are shifting it by one whole period. If we shift it by half of this, pi, then we will have a wave completely opposite to what we had. To really get a better grip on this, nothing better than trying it out:

```python
amplitude = 5
frequency = 5
phase_difference = numpy.pi/4
y = amplitude*numpy.cos(omega*t + phase_difference )
z = amplitude*numpy.cos(omega*t)
pyplot.plot(t,y)
pyplot.plot(t,z)
pyplot.xlabel("Time")
pyplot.ylabel("Amplitude")
pyplot.title('Phase difference is the shift in time of the wave')
```

## Wave interference

Waves meet and interfere. Fortunately for us, they interfere in a very simple way: we can just add up their amplitudes. This is called the principle of superposition and it is illustrated below. If you run the following code, you will get the sum result of the two waves we just looked at in the phase difference section.

```python
sum = z + y
pyplot.plot(t, sum)
```

In the following code, you can change the amplitude, frequency and phase difference of either wave to see the effect on the total wave

```python
a_amplitude = 3
b_amplitude = 5
a_frequency = 5
b_frequency = 7
a_phase_difference = 0
b_phase_difference = numpy.pi
a = a_amplitude*numpy.cos(a_frequency*2*numpy.pi*t+a_phase_difference)
b = b_amplitude*numpy.cos(b_frequency*2*numpy.pi*t+b_phase_difference)
c = a + b
pyplot.plot(t,a)
pyplot.plot(t,b)
pyplot.plot(t,c)
```
